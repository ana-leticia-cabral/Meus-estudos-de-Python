1) Faça um programa que receba uma frase e mostre a frase maiúscula e a quantidade de caracteres em branco.

```python
def info_frase():
    frase = input("Digite uma frase: ")
    frase_maiuscula = frase.upper()
    caracteres_branco = frase.count(" ")
    print(f"Frase maiúscula: {frase_maiuscula}\n"
          f"Quantidade de caracteres em branco: {caracteres_branco}.")

info_frase()
```

2) Faça um programa para calcular e mostrar a maior e menor altura de um grupo de 10 pessoas, a média de altura das mulheres e a quantidade de homens.


```python
def info_10_pessoas():
    lista_nome = []
    lista_altura = []
    lista_sexo = []
    contador = 0

    while len(lista_nome) < 10:
        nome = input(f'Informe o nome da pessoa {contador + 1}: ')
        altura = float(input('Informe a altura (maior que 0): '))
        sexo = input("Informe o sexo (M/F): ").upper()

        nome_valido = all(palavra.isalpha() for palavra in nome.split())
        sexo_valido = sexo in ('M', 'F')

        if nome_valido and altura > 0 and sexo_valido:
            lista_nome.append(nome)
            lista_altura.append(altura)
            lista_sexo.append(sexo)
            contador += 1
        else:
            print("Dados inválidos. Tente novamente.")


    informacoes_pessoas = list(zip(lista_nome, lista_altura, lista_sexo))

    quantidade_mulheres = 0
    altura_mulheres = 0
    quantidade_homens = 0

    for nome, altura, sexo in informacoes_pessoas:
        if sexo == 'F':
            quantidade_mulheres += 1
            altura_mulheres += altura
        elif sexo == 'M':
            quantidade_homens += 1

    maior_altura = max(lista_altura)
    menor_altura = min(lista_altura)
    media_altura_mulheres = altura_mulheres / quantidade_mulheres if quantidade_mulheres > 0 else 0

    for i, (nome, altura, sexo) in enumerate(informacoes_pessoas, 1):
        print(f'Pessoa {i}: {nome}, {altura}m, {sexo}')

    print(f'Maior altura: {maior_altura}\n'
          f'Menor altura: {menor_altura}\n'
          f'Média de altura das mulheres: {media_altura_mulheres:.2f}\n'
          f'Quantidade de homens: {quantidade_homens}')


info_10_pessoas()


```

3) Faça um programa que receba uma letra maiúscula ou minúscula, e mostre todas as letras do alfabeto entre a letra fornecida e a última letra. Exemplo: Entrada = "g"
Saída = "h,i,j,k...z"

```python
def alfabeto():

  alfabeto = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y', 'z']

  letra = input('Digite uma letra: ')

  while letra.lower() not in alfabeto:
    print('Dados inválidos. Digite novamente.')
    letra = input("Digite uma letra: ")


  indice = alfabeto.index(letra.lower())

  resto_alfabeto = [l for l in alfabeto[indice + 1:]]

  print('Restante do alfabeto: ' +', '.join(resto_alfabeto))


alfabeto()
```

4) Faça um sistema para cadastrar funcionários e mostrar o maior, o menor salário e a média salarial. A entrada é composta pelo código do funcionário e seu salário.


```python
funcionarios = {}

def funcionario(codigo_func, salario):
  funcionarios[codigo_func] = salario

  print(f'Funcionários cadastrados: {funcionarios}\n'
        f'Maior salário cadastrado: {max(funcionarios.values()):.2f}\n'
        f'Menor salário cadastrado: {min(funcionarios.values()):.2f}\n'
        f'Média salarial: {sum(funcionarios.values())/len(funcionarios):.2f}')
```
